# 데이터 프레임 불러오기

csv 파일 불러오기 : read_csv()

tsv 파일은 따로 불러오는 함수가 없기 때문에 read_csv() 함수의 매개변수를 조정해주어야 함

```python
import pandas as pd
```

## tsv 파일 불러오기
* \t : Tab키를 의미함
* sep 매개변수를 통해 데이터를 구분하는 기준을 설정할 수 있음

```python
df = pd.read_csv("gapminder.tsv", sep='\t')
df
```

## 열에 접근하기

```python
df['country']
```

```python
df[['country', 'year']]
```

## 행에 접근하기
* loc : 이름으로 접근
* iloc : 순서로 접근

[행,열]로 접근 가능

```python
df.loc[0]
```

```python
df.iloc[0]
```

```python
df.loc[[0,10,100,1000]]
```

```python
df.iloc[[0,10,100,1000]]
```

```python
df.loc[[0,10,100,1000], ['country', 'year']]
```

```python
df.iloc[[0,10,100,1000], [0, 2]]
```

## 원하는 행만 추출하는 방법
year의 값이 year의 평균값보다 큰 행만 추출하기

```python
df[df['year'] > df['year'].mean()]
```

year의 값이 year의 평균값보다 크고 동시에 lifeExp값도 lifeExp값의 평균보다 큰 행만 추출하기

```python
df[(df['year'] > df['year'].mean()) & (df['lifeExp'] > df['lifeExp'].mean())]
```

year의 값이 year의 평균값보다 크거나 또는 lifeExp값도 lifeExp값의 평균보다 큰 행만 추출하기

```python
df[(df['year'] > df['year'].mean()) | (df['lifeExp'] > df['lifeExp'].mean())]
```
